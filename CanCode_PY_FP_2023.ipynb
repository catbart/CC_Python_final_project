This project reads in raw demographic data for AP Computer Science (CS) Exams taken and pass rates for years 2008-2013 from a CSV file and stores the information in a Pandas DataFrame. The raw data was compiled by Barbara Ericson from College Board Data and made available for public use on data.world. The data set is broken down by all, female, black and hispanic students taking the exam in a given year, state breakouts, and pass rates. It also includes data on the number of AP CS courses offered and AP Calculus courses offered. The following code captures segments of that data to create pictorial representations of demographic disparities of AP CS enrollment and pass rates using MatPlotLib graphing applications. 

```python
"""This code section reads in AP Computer Science data for years 2008-2013 
  from a CSV file (https://data.world/education/ap-cs-pass-rate-by-racegender/workspace/file?filename=pass_rate.csv), 
  separates the data into two files; 1. containing data on states collectively and, 2. one containing overall national data. 
  For simplicity in labeling, the number of students 'enrolled' was assumed to be the number of students 'taking the AP exam' for which data actual data was provided. """

# Import libraries for data handling and graphing
import pandas as pd
from matplotlib import pyplot as plt
import numpy as np 

# Read raw data from CSV file into GitHub repository
df = pd.read_csv('https://raw.githubusercontent.com/catbart/CC_Python_final_project/main/CS_by_Gender_Race_08_13.csv', na_values='_' )
# Create dataframe, 'data', containing individual states' AP Comp Sci data for years 2008-2013
data = df[14:]
# Create dataframe, 'nat_cs_stat', containing national AP Comp Sci data for years 2008-2013
nat_cs_stat = df[0:6:]
# Create dataframe , 'nat_calc_stat', containing national AP Calculus data for years 2008-2013
nat_calc_stat = df[7:12:]
```

```python
# Create python function to compute mean, median, max & min for State AP CS data colummns called in dataframe 'data'

def data_stats(dat_col):
  print(dat_col)

  print('Mean:', data[dat_col].mean())
  print('Median:', data[dat_col].median())
  print('Maximum:', data[dat_col].max())
  print('Minimum:', data[dat_col].min())
  print('\n')
```

```python
# Create python function to compute mean, median, max & min for National AP CS data colummns called in dataframe 'nat_cs_stat'

def nat_cs(dat_coln):
  print(dat_coln)

  print('Mean:', nat_cs_stat[dat_coln].mean())
  print('Median:', nat_cs_stat[dat_coln].median())
  print('Maximum:', nat_cs_stat[dat_coln].max())
  print('Minimum:', nat_cs_stat[dat_coln].min())
  print('\n')
```

```python
# Call functions to compute and print statistics for various columns of dataframe called

# Compute statistics on individual state AP CS datasets
print("All states, years 2008-2013 AP CS Statistics",'\n')
data_stats("Pct passed")
data_stats("Pct female passed")
data_stats("Pct Black passed")
data_stats("Pct Hispanic passed")
data_stats("yield per teacher")

# Compute statistics on National AP CS datasets
print('Total Number of AP CS students Nationwide: ','\n')
nat_cs("Total Num")
nat_cs("Pct passed")
nat_cs("Pct female passed")
nat_cs("Pct Black passed")
nat_cs("Pct Hispanic passed")
nat_cs("yield per teacher")
```

```python
# Create dataset: Num Schools Offering AP Calc or AP CS in Year 2013
APcourse = ['nat_cs_stat', 'nat_calc_stat'] 
datanat = [nat_cs_stat['Num schools offering'][5], nat_calc_stat['Num schools offering'][11]]

# Create Pie Chart depicting AP CS courses offered as compared to AP Calc courses offered

  # Set Figure Size
fig = plt.figure(figsize =(15, 10))
  # Select data source with labels; set font size and chart properties
plt.pie(datanat, labels = ['AP Comp Sci','AP Calc'], textprops={'fontsize': 18}, explode = (0.0, 0.1))
  # Add Title and set title fontsize
plt.title("Fig. 1: Schools offering AP CS as compared to AP Calc ", fontsize=20)
  # show plot
plt.show()
```

```python
# Create plot showing number of AP Computer Science courses offered each year from 2008-2013
df.iloc[0:6].plot(x = 'Year', y='Num schools offering', linewidth=10)
plt.xlabel("Year", fontsize=18)
plt.ylabel("Num Schools Offering", fontsize=18)
plt.title("Fig. 2: Num Schools Offering AP CS", fontsize=18)
plt.show()

# Create plot showing number of AP Calculus courses offered each year from 2008-2013
plt.plot(df.iloc[6:12]['Year'],df.iloc[6:12]['Num schools offering'], color='orangered', linewidth=10)
plt.xlabel("Year", fontsize=18)
plt.ylabel("Num Schools Offering", fontsize=18)
plt.title("Fig. 3: Num Schools Offering AP Calc", fontsize=18)
plt.show()
```

```python
# Create dataset: Num Females vs Males who took AP CS in Year 2013
AP_CS_Gender = ['nat_cs_stat', 'nat_calc_stat'] 
datanat = [(nat_cs_stat['Total Num'][5]-nat_cs_stat['Num female'][5]), nat_cs_stat['Num female'][5]]

# Create plot
fig = plt.figure(figsize =(15, 10))
plt.pie(datanat, labels = ['AP CS Male ','AP CS Female'],textprops={'fontsize': 18}, explode = (0.0, 0.1), colors = ['skyblue','pink'])
plt.title("Fig. 4: Num Males compared to Num Females taking AP CS in 2013 ", fontsize=20)
 
# show plot
plt.show()
```

```python
# Create histograms showing distribution of Female and All Student AP CS pass rates

fig, axs = plt.subplots(1,2, figsize = (15,5))
# Create first histogram, showing distribution pass rates for female AP CS students, years 2008-2013
plt.subplot(1,2,1)
plt.hist(nat_cs_stat['Pct female passed'], color = "pink")
# adding labels
plt.xlabel('Score Distribution')
plt.ylabel('Distribution of 2008-2013 pass rates')
plt.title('Fig. 5: Distribution of AP CS Female Pass Rates for 2008-2013')

# Create second histogram, showing distribution pass rates for all AP CS students, years 2008-2013
plt.subplot(1,2,2)
plt.hist(nat_cs_stat['Pct passed'], color = 'skyblue')
# adding labels
plt.xlabel('Score Distribution')
plt.ylabel('Distribution of 2008-2013 pass rates')
plt.title('Fig. 6: Distribution of All AP CS Student Pass Rates for 2008-2013')
```

```python
# Create graph showing male & female AP CS Enrollment in 2013 for states

# Create dataframe containing state data for year 2013
data2013 = data[data['Year']==2013]

# Set bar graph parameters
X_axis = np.arange(len(data2013))
fig, axs = plt.subplots( figsize = (10,10))
ax = plt.subplot(111)
w = 0.3
# Calculate number of males enrolled as total number - female number
ax.bar(X_axis - w, (data2013['Total Num']-data2013['Num female']), width=w, color='skyblue', align='center', label = 'Male AP CS Enrollment')
ax.bar(X_axis, data2013['Num female'], width=w, color='pink', align='center', label = 'Female AP CS Enrollment')

# Set x-axis labels, rotating state names 
plt.xticks(X_axis, data2013['State'], rotation = 90)

# Define graph labels
plt.xlabel("State", fontsize=18)
plt.ylabel("Number Enrolled", fontsize=18)
plt.title("Fig. 7: Male & Female AP Comp Sci Enrollment by State in 2013", fontsize=18)
plt.legend()

# Show Plot
plt.show()
```

```python
# Create graph to show percent Females Passing vs All Student Passing for Year 2013
# Data set for year 2013, was previously created above

 # Set bar graph parameters, bar size, offset, source, label & colors
plt.bar(X_axis - 0.2, data2013['Pct passed'], 0.4, label = 'All AP CS Students Pct Passed', color = 'grey')
plt.bar(X_axis + 0.2, data2013['Pct female passed'], 0.4, label = 'Female AP CS Pct Passed', color = "pink")

# Set figure size
plt.rcParams['figure.figsize'] = (10, 10)  

# Define axis labels, font size, and Title
plt.xticks(X_axis, data2013['State'], rotation = 90)
plt.xlabel("State", fontsize=18)
plt.ylabel("Percent Passed", fontsize=18)
plt.title("Fig. 8: Female as Compared to All Students CS Pass Rate in 2013", fontsize=18)

# Show legend and Plot
plt.legend()
plt.show()
```

```python
# Create dataset: Num Black & Hispanic who took AP CS in Year 2013
AP_CS_Race = ['nat_cs_stat', 'nat_calc_stat'] 
datanat = [(nat_cs_stat['Total Num'][5]-nat_cs_stat[' Num Black'][5]-nat_cs_stat['Num Hispanic'][5]), nat_cs_stat[' Num Black'][5],nat_cs_stat['Num Hispanic'][5]]

# Create plot
fig = plt.figure(figsize =(15, 10))
plt.pie(datanat, labels = ['AP CS Not Black or Hispanic ','AP CS Black', 'AP CS Hispanic'],textprops={'fontsize': 18}, colors = ['purple','g','r'])
plt.title("Fig. 9: Number Black or Hispanic relative to ALL students taking AP CS in 2013 ", fontsize=20)
 
# show plot
plt.show()
```

```python
# Create two plots showing number of Blacks & Hispanics relative to students NOT Black or Hispanic enrolled in AP CS in 2013
plt.subplots(figsize=(25, 7.5))

# Create sub plot of AP CS Enrollment by available race data
ax = plt.subplot(1,2,1)
w = 0.3
ax.bar(X_axis - w, (data2013['Total Num']-data2013[' Num Black']-data2013['Num Hispanic']), width=w, color='purple', align='center', label = 'Not Black or Hispanic AP CS Enrollment')
ax.bar(X_axis, data2013[' Num Black'], width=w, color='g', align='center', label = 'Black AP CS Enrollment')
ax.bar(X_axis + w, data2013['Num Hispanic'], width=w, color='r', align='center', label = 'Hispanic AP CS Enrollment')
# Define plot labels & Title 
plt.xticks(X_axis, data2013['State'], rotation = 90)
plt.xlabel("State", fontsize=18)
plt.ylabel("Number Enrolled", fontsize=18)
plt.title("Fig. 10: AP CS Black & Hispanic Students Enrolled in 2013", fontsize=20)
plt.legend()

# Create sub plot of AP CS pass rates by available race data
ax = plt.subplot(1,2,2)
w = 0.3
ax.bar(X_axis - w, data2013['Pct passed'], width=w, color='gray', align='center', label = 'All AP CS Students Pct Passed')
ax.bar(X_axis, data2013['Pct Black passed'], width=w, color='g', align='center', label = 'Black AP CS Pct Passed')
ax.bar(X_axis + w, data2013['Pct Hispanic passed'], width=w, color='r', align='center', label = 'Hispanic AP CS Pct Passed')
# Define plot labels & Title  
plt.xticks(X_axis, data2013['State'], rotation = 90)
plt.xlabel("State", fontsize=18)
plt.ylabel("Percent Passed", fontsize=18)
plt.title("Fig. 11: Percent Black & Hispanic Students Passed in 2013", fontsize=20)
plt.legend()

# Show Plots
plt.show()
```
